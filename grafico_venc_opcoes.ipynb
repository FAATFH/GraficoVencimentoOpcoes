```python
def e_vencimento(s):
    s = str(s)[0:10]
    if s in vencimentos:
        return True

cotacoes['venc'] = cotacoes.apply(lambda row : e_vencimento(row['Date']), axis = 1)
```

```python
df = cotacoes.loc[cotacoes['venc'] == True]

def close(a, b):
    c = cotacoes.loc[(cotacoes['Date'] >= a) & (cotacoes['Date'] <= b)]
    close = c['Close'].iloc[-2]
    return close

def high(a, b):
    c = cotacoes.loc[(cotacoes['Date'] >= a) & (cotacoes['Date'] <= b)]
    high = c['High'].max()
    return high

def low(a, b):
    c = cotacoes.loc[(cotacoes['Date'] >= a) & (cotacoes['Date'] <= b)]
    low = c['Low'].min()
    return low

df['next_exp'] = df['Date'].shift(-1)
df.drop(df.tail(1).index,inplace=True)

df['candle_close'] = df.apply(lambda row : close(row['Date'], row['next_exp']), axis = 1)
df['candle_high'] = df.apply(lambda row : high(row['Date'], row['next_exp']), axis = 1)
df['candle_low'] = df.apply(lambda row : low(row['Date'], row['next_exp']), axis = 1)

df
```
